### _1._ (Section 1.1)  

Solve exactly (by hand) the following differential equations:

  (a) $dy/dt = 3y^2$      
  (b) dy/dt = 2+4y      
  (c) dy/dt =  7t.

Use the condition : y(t=0) = 1. 


##

### _1a._ 

$dy/dt = 3y^2$

$\int y^-2\, dy = \int 3\, dt$

$-y^-1 = -3t + C$

$y = \frac{1}{-3t + C}$

$ $

since y(0) = 1, then:

$1 = \frac{1}{-3(0) + C} = \frac{1}{C}$

$C = 1$

$$=>  y = \frac{1}{-3t + 1}$$


$ $

### _1b._ 

$\frac{dy}{dt} = 2 + 4y$

$dy = (2 + 4y)\, dt$

$dy = 2\,dt + 4y\,dt$

$dy = 2(1 + 2y)\,dt$

$\int\frac{1}{1 + 2y}\,dy = \int 2\,dt$

$\frac{1}{2}\int\,[ln(1 + 2y)]'\,dy = \int 2\, dt$

$\frac{1}{2}ln(1 + 2y) = 2t + C$

taking the exponent of both sides:

$1 + 2y = e^{4t+C}$

$1 + 2y = Ce^{4t}$

$2y = Ce^{4t} - 1$

$y = Ce^{4t} - \frac{1}{2}$

since y(0) = 1, then:

$1 = Ce^{0} - \frac{1}{2}$

$C = 1 + \frac{1}/{2}$

$C = \frac{3}{2}$

$$=> y = \frac{3}{2}e^{4t} - \frac{1}{2}$$


### _1c._ 

$dy/dt = 7t$

$dy = 7t\,dt$

$\int dy = \int 7t\, dt$

$y = \frac{7}{2}t^{2} + C$

since y(0) = 1, then:

$1 = \frac{7}{2}(0)^{2} + C$

$C = 1$

$$=> y = \frac{7}{2}t^{2} + 1$$

## 3.	(Section 1.1.2)  

Use Program 1.2  (rewritten in Python) to plot the Euler’s method approximation to the solution to:  dy/dt = sqrt(y) with the following parameters:
a.	Initial condition:  y[0] = 5;
b.	Time step = 0.1
c.	Time interval: [0,4]
Plot also the analytical solution. You don’t need to label the points as in Figure 1.3—instead, use the legend()command.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


t = 0
y = 5
dt = 0.1
t_final = 4
N_steps = round(t_final / dt) + 1
t_vec = np.linspace(t, t_final, N_steps)
y_euler_vec = [y]

for i in np.arange(len(t_vec)-1):
    y_new = y + (dt * np.sqrt(y))
    y_euler_vec.append(y_new)
    y = y_new
    
y_true = ((np.array(t_vec) + np.sqrt(20))**2)/4


    
plt.xlabel('t')
plt.ylabel('y')
plt.xlim([0,5])
#plt.ylim([0,1])
#plt.yticks([5])
plt.grid(color = 'g', linestyle = 'dotted')
plt.plot(t_vec, y_euler_vec, '*--', label = 'Euler')
plt.plot(t_vec, y_true, 'r', label = 'y = $e^{-t}$')
plt.title('')
plt.legend()

plt.show()

print(N_steps)
```

Compute  (by hand) the Taylor series for:  y = $7e^{0.6t}$. Duplicate Figure 1.4 using this function instead: that is, plot the function, and Taylor series approximations with 2,3 4,5 terms.  Use the interval from t=0 to t=3. To label the functions, use the text command from matplotlib (look it up: see https://matplotlib.org/users/text_intro.html)


$f^{(0)}(t) = 7e^{0.6t}$

$f'(t) = 4.2e^{0.6t}$

$f"(t) = 2.52e^{0.6t}$

$f^{(3)}(t) = 1.512e^{0.6t}$

$f^{(4)}(t) = 0.9072e^{0.6t}$

$f^{(5)}(t) = 0.54432e^{0.6t}$

### At t = 0:

$f^{(0)}(0) = 7$

$f'(0) = 4.2$

$f"(0) = \frac{2.52}{2}$

$f^{(3)}(0) = \frac{1.512}{3!}$

$f^{(4)}(0) = \frac{0.9072}{4!}$

$f^{(5)}(0) = \frac{0.54432}{5!}$


### Taylor series of $y = 7e^{0.6t}$ centered at the origin:

$y = f(t) = f(0) + f'(0)\,t + \frac{f"(0)}{2!}\,t^{2} + \frac{f^{(3)}(0)}{3!}\,t^{3} + \frac{f^{(4)}(0)}{4!}\,t^{4} + ...$

$y = f(t) = 7 + 4.2t + \frac{2.52}{2!}\,t^{2} + \frac{1.512}{3!}\,t^{3} + \frac{0.9074}{4!}\,t^{4} + ...$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 3)

y = 7 * np.exp(0.6*t)

y_2 = 7 + 4.2*t

y_3 = y_2 + (2.52/2)*(t**2)

y_4 = y_3 + (1.512/6)*(t**3)

y_5 = y_4 + (0.9074/24)*(t**4)


    
plt.xlabel('t')
plt.ylabel('y(t)')
plt.xlim([0,3])
# #plt.ylim([0,1])
# #plt.yticks([5])
plt.grid(color = 'g', linestyle = 'dotted')
plt.plot(t, y, label = 'y = $7e^{0.6t}$')
plt.plot(t, y_2, '--', label = 'y_2')
plt.plot(t, y_3, '--', label = 'y_3')
plt.plot(t, y_4, '--', label = 'y_4')
plt.plot(t, y_5, '--', label = 'y_5')
plt.text(3, 42, r'$7e^{0.6t}$')
plt.text(3, 40, r'$7 + 4.2t + \frac{2.52}{2}t^{2} + \frac{1.512}{6}t^{3} + \frac{0.9074}{24}t^{4}$')
plt.text(3, 37, r'$7 + 4.2t + \frac{2.52}{2}t^{2} + \frac{1.512}{6}t^{3}$')
plt.text(3, 30, r'$7 + 4.2t + \frac{2.52}{2}t^{2}$')
plt.text(3, 19, r'$7 + 4.2t$')
plt.title('')
#plt.legend()

plt.show()

```
