## Mandelbrot set

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


###radius is given as 0.25 and we want to exam of height of 0.1 extending from the
###section of the circumference that is being evaluated. Therefore, r_max = 0.25 + 0.1 = 0.35
r_min = 0.25 ##minimum radius value
r_max = 0.35 ## max radius
theta_min = (2 * np.pi / 3)  ##minimum theta value
theta_max = 5 * np.pi / 6
index_max = 64
n_r = 101 ##number of values of r
n_theta = 151 #3 no. of values of theta


###want to plot (r, theta, index(-1 + r cos⁡θ+i  r sin⁡θ))

## Array to contain n_r evenly spaced values of r between r_min and r_max
rVals = np.linspace(r_min, r_max, n_r)


##  Array to contain n_theta evenly spaced values of theta between theta_min and theta_max
thVals = np.linspace(theta_min, theta_max, n_theta)


## Array with n_r rows and n_theta columns to contain the index values for different (r,theta) pairs.
indexVals = np.zeros((n_r, n_theta))



##Loop through r values, theta values and index...
## NB 1j is the symbol used to represent 'i' in complex numbers
for ii in range(n_r):
    for jj in range(n_theta):
        c = (-1 + rVals[ii] * np.cos(thVals[jj])) + 1j * (rVals[ii]*np.sin(thVals[jj]))
        z = c

        indexVals[ii, jj] = index_max

        for kk in range(index_max):
            z = z**2 + c
            
            if abs(z) > 2:
                indexVals[ii, jj] = kk + 1 ##kk + 1 since python has 0 indexing. So, an element with index 1 would be the second element in the array
                
                break
plt.figure(1)                
plt.figure(figsize = (10, 9))
plt.imshow(indexVals, cmap = 'jet') #.cmap(indexVals)
plt.colorbar()
plt.xlabel('r', fontsize = 12)
plt.ylabel('theta', fontsize = 12)
plt.title('Mandelbrot Plot: $120^o < \Theta < 150^o$', fontsize = 18)
#plt.tight_layout()


plt.figure(2)






###gca?? "Get the current Axes instance on the current figure matching the given keyword args, or create one." 
### that is get the current axes on the current vfigure
plt.gca().invert_xaxis()

plt.gca().invert_yaxis()


```
