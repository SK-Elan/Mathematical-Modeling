```python
### 8
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#100 evenly spaced values of theta from 0 to 2pi
th_Vals = np.linspace(0, 2*np.pi, 100)

x = np.cos(th_Vals)
y = np.sin(th_Vals)

#The equation for c was determined to be
c = (0.5*x - (0.25*(x**2 - y**2))) + 1j*(0.5*y - 0.5*x*y)

xVals = np.real(c)
yVals = np.imag(c)

plt.plot(xVals, yVals, '*')
plt.xlabel('Real c')
plt.ylabel('Im c')
plt.title('c = x + iy')
```

```python
##11

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

######################################### |c+1| = 0.25

###########
#Now suppose c+1 is a vector of complex numbers instead of a single vector
# c+1 is a circle
n = 16
m = 3

#Three points in circle with period 3
cVec = np.array([-.08 + 0.73j, -0.12 + 0.75j, -0.16 + 0.75j])


##initialize vector to store terms of the Mandelbrot sequence
m_arr = np.zeros((n, m)) * 1j
m_arr[0] = cVec  #cVec is the first row of m_arr

##Do a for loop to update the remaining rows of m_arr
for ii in range(1, n):
    m_arr[ii,:] = (m_arr[ii-1, :])**2 + cVec


arr2 = m_arr[0::3,:]
arr3 = m_arr[1::3,:]
arr4 = m_arr[2::3,:]

    
###AIMING TO PLOT ALL TERMS OF THE MANDELBROT SEQUENCE IN FIGURE 1   
xVals = np.real(m_arr)
yVals = np.imag(m_arr)

###AIMING TO PLOT Z0, Z3, Z6,... IN FIGURE 2
x2 = np.real(arr2)
y2 = np.imag(arr2)


###AIMING TO PLOT Z1, Z4, Z7,... IN FIGURE 2
x3 = np.real(arr3)
y3 = np.imag(arr3)


###AIMING TO PLOT Z1, Z4, Z7,... IN FIGURE 2
x4 = np.real(arr4)
y4 = np.imag(arr4)




##CREATE AN INSTANCE OF plt.figure()
plt.figure(figsize = (16,8)) #figsize = (8,8), num = 1, 


###Figure 1
#plt.figure(num = 2, figsize = (8,8)) #figsize = (8,8)
plt.subplot(2,2,1)
plt.plot(xVals, yVals, '*')
plt.title('All terms, Period 3')
plt.xlabel('Real')
plt.ylabel('Imag')


###Figure 2
plt.subplot(2,2,2)
plt.plot(x2, y2, '*')
plt.title('Z0, Z3, Z6,...')
plt.xlabel('Real')
plt.ylabel('Imag')


###Figure 3
plt.subplot(2,2,3)
plt.plot(x3, y3, '*')
plt.title('Z1, Z4, Z7,...')
plt.xlabel('Real')
plt.ylabel('Imag')


###Figure 4
plt.subplot(2,2,4)
plt.plot(x4, y4, '*')
plt.title('Z2, Z5, Z8,...')
plt.xlabel('Real')
plt.ylabel('Imag')



plt.tight_layout()


print(" ")
print(" ")
print(" ")

print(" ")
print(" ")
print(" ")










```
