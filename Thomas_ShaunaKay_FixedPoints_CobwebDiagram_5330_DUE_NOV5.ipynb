j)	Write a function that takes x, a and n as inputs, which gives the value of $F^{n}(x)$ where F is given by Equation (3).  Use your function to plot the function Plot $F^{3}(x)$ corresponding to a = 3.8, 3.8284, and 3.8495 (put the 3 functions on the same plot).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


#Function to calculate the nth iterate value of f, in the non-linear difference equation
def fn(x, a, n):
    xtPlus1 = a * x *(1-x) # Initialize xtPlus1 to substitute later in formula
    
    fVal = xtPlus1
    
    for i in range(1, n):
        fVal = a * fVal * (1 - fVal) # iterating n times
#         xtPlus1 = fVal #update xtPlus1
        
    return fVal

# fn(0.5, 1, 3)


#Use your function to plot the function Plot F(3)(x) corresponding to a = 3.8, 3.8284, and 3.8495 
#(put the 3 functions on the same plot).

a = np.array([3.8, 3.8284, 3.8495]) #Array with varying values of 'a'

n = 3 #Number of iterations

xVals = np.arange(0, 1, 0.01) #Range for x Values

yVals = np.zeros((len(a), len(xVals))) # 2d array(matrix) to store the values for y for varying values of 'a'

for j in range(len(a)): #Loop through each 'a'
    for i in range(len(xVals)): #Loop through each x in order to calculate its corresponding y value.
        yVals[j,i] = fn(xVals[i], a[j], n)

    
# print('yVals:', yVals[0:,], '\n')
# print(yVals.shape)

# print('xVals:', xVals[0:,], '\n')

#####Plotting each function on the same plot
plt.figure(figsize = (12,10))

plt.plot(xVals, yVals[0,:], 'b-', label = 'a = 3.8', linewidth = 1)
plt.plot(xVals, yVals[1,:], 'r-', label = 'a = 3.8284', linewidth = 1)
plt.plot(xVals, yVals[1,:], 'g*', label = 'a = 3.8495', linewidth = 2)

plt.xlabel('$X_t$', fontsize = 14)
plt.ylabel('$X_{t+3}$', fontsize = 14)
plt.title('')



plt.legend()
plt.show()


        

```
